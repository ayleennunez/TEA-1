```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("FAOSTAT_data_es_9-21-2022.csv")
data
```

```python
data = data.groupby(["Área","Año"], sort=False)["valor"].max()
data.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1)
```

```python
datos2 = pd.read_csv("Datos_TEA.csv")
datos2
```

```python
cereales = datos2[datos2.Productos =="Cereales"]
cereales
```

```python
ganado = datos2[datos2.Productos =="Ganado"]
ganado
```

```python
data1 = datos2.groupby(["Productos","Year"], sort=False)["Valor"].max()
data1.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1)
```

```python
data2 = ganado.groupby(["Productos","Year"], sort=False)["Valor"].max()
data2.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=1)
```
